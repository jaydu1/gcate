```python
sns.set(font_scale=1.5, style='white')


cell_types = ['T4', 'cM', 'B', 'T8', 'NK']

for cell_type in cell_types:
    print(cell_type)

    file_name = path_data + 'data_lupus_{}.npz'.format(cell_type)
    with open(file_name, 'rb') as f:
        tmp = np.load(f)
        X = tmp['X']
        Y = tmp['Y'].astype(float)
        X = X * 2 - 1
        X = np.c_[np.ones(X.shape[0]), np.log(np.sum(Y, axis=1)), X]
        X = X / np.sqrt(np.sum(X**2, 0, keepdims=True)) * np.sqrt(X.shape[0])
    X_sub = np.c_[X[:,:2],X[:,-1:]]
    id_genes = np.any(Y, axis=0) & (np.sum(Y>0, axis=0)>=10)
    p_all = Y.shape[1]
    Y = Y[:,id_genes]
    print(X.shape, X_sub.shape, Y.shape)

    p = Y.shape[1]
    d = X.shape[1]

    with open('result/lupus/dill_{}.pkl'.format(cell_type), "rb") as file_object:
        [
            B_glm,tvals,pvals,_df_glm,disp_glm,
            B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
            B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
            B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
        ] = dill.load(file_object)

    kwargs_glm = {
            'family':'negative_binomial',
            'nuisance':1/disp_glm_sub[None,:]
        }

    r_max = 10
    df_r = estimate_r(Y, X_sub, r_max, 
        kwargs_glm=kwargs_glm, 
        kwargs_ls_1={'alpha':0.1}, kwargs_es_1={'max_iters': 2000},
        kwargs_ls_2={'alpha':0.01}, kwargs_es_2={'max_iters': 2000, 'patience':25},
        c1=0.01, intercept=1, offset=0, C=1e3, verbose=False
    )

    
    c = 0.2 if cell_type == 'NK' else 0.25
    fig = plot_r(df_r, c=c)
    fig.savefig(path_result + 'lupus_{}_r.pdf'.format(cell_type), bbox_inches='tight')    
```

## Figure G6: Selection of penalty parameter

```python
mad_max_dict = {'T4':1.13, 'cM':1.13, 'B':1.13, 'T8':1.11, 'NK':1.}
med_max_dict = {'T4':.025, 'cM':.025, 'B':.025, 'T8':.08, 'NK':.08}

dict_disc = {}
for cell_type in cell_types[:]:
    file_name = path_data + 'data_lupus_{}.npz'.format(cell_type)
    with open(file_name, 'rb') as f:
        tmp = np.load(f, allow_pickle=True)
        X = tmp['X']
        Y = tmp['Y'].astype(float)
        gene_names = tmp['gene_names'].astype(str)
        X = X * 2 - 1
        X = np.c_[np.ones(X.shape[0]), np.log(np.sum(Y, axis=1)),  X]
        X = X / np.sqrt(np.sum(X**2, 0, keepdims=True)) * np.sqrt(X.shape[0])
        id_genes = np.any(Y!=0, axis=0) & (np.sum(Y>0, axis=0)>=10)
        
        Y = Y[:,id_genes]
        gene_names = gene_names[id_genes]

    n = Y.shape[0]
    X_sub = np.c_[X[:,:2],X[:,-1:]]

    print(cell_type)
    print('\t shape:', X.shape, X_sub.shape, Y.shape)
    print('\t sparsity:',np.mean(Y==0))

    
    with open(path_result+'dill_{}.pkl'.format(cell_type), "rb") as file_object:
        [
            B_glm,tvals,pvals,_df_glm,disp_glm,
            B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
            B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
            B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
        ] = dill.load(file_object)

    df_glm_nb = pd.DataFrame({
        'beta_hat':B_glm_nb[:,-1],
        'z_scores':tvals_glm_nb, 
        'p_values':pvals_glm_nb,
        'q_values':multipletests(pvals_glm_nb, alpha=0.05, method='fdr_bh')[1]
    })
    df_glm_nb_sub = pd.DataFrame({
        'beta_hat':B_glm_nb_sub[:,-1],
        'z_scores':tvals_glm_nb_sub, 
        'p_values':pvals_glm_nb_sub,
        'q_values':multipletests(pvals_glm_nb_sub, alpha=0.05, method='fdr_bh')[1]
    })

    df_gcate = pd.read_csv(path_result+'gcate_{}.csv'.format(cell_type))
    fig, (lams, medians, mads) = plot_lam(df_gcate, 
        mad_max=mad_max_dict[cell_type], med_max=med_max_dict[cell_type])
    fig.savefig(path_result + 'lupus_{}_lam.pdf'.format(cell_type), dpi=300, bbox_inches='tight')
    idx = (np.abs(medians) < med_max_dict[cell_type])&(mads < mad_max_dict[cell_type])
    lam = lams[idx][np.argmax(np.abs(mads[idx]))]

    print(lam, lam/np.sqrt(np.log(n)/n))
    df_gcate = df_gcate[df_gcate['lam']==lam].reset_index(drop=True)
    df_cate = pd.read_csv('{}cate_raw_{}.csv'.format(path_result, cell_type))
    df_cate_adj = pd.read_csv('{}cate_{}.csv'.format(path_result, cell_type))

    print('\t GLM discoveries:', np.sum(df_glm_nb['p_values']<0.05), np.sum(df_glm_nb['q_values']<0.2))

    pvals = np.c_[df_glm_nb['p_values'], df_glm_nb_sub['p_values'], df_cate['p_values'],
        df_cate_adj['p_values'], df_gcate['p_values']]
    discoveries = (pvals<0.05)
    TP = (discoveries[:,1:] == 1) & (discoveries[:,:1] == 1)
    FP = (discoveries[:,1:] == 1) & (discoveries[:,:1] == 0)
    print('\t p:', np.sum(TP,axis=0), np.sum(FP,axis=0))

    qvals = np.c_[df_glm_nb['q_values'], df_glm_nb_sub['q_values'], df_cate['q_values'],
        df_cate_adj['q_values'], df_gcate['q_values']]
    discoveries = (qvals<0.2)
    TP = (discoveries[:,1:] == 1) & (discoveries[:,:1] == 1)
    FP = (discoveries[:,1:] == 1) & (discoveries[:,:1] == 0)    
    print('\t q:', np.sum(TP,axis=0), np.sum(FP,axis=0))
    sns.set_theme(style="white", font_scale=1.2)
    fig, axes = plt.subplots(2,3,figsize=(10,6), sharex=False)

    axes[0,0].hist(df_glm_nb['z_scores'], bins=50, density=True, range=[-10,10])
    axes[0,1].hist(df_glm_nb_sub['z_scores'], bins=50, density=True, range=[-10,10])
    axes[0,2].hist(df_cate['z_scores'], bins=50, density=True, range=[-10,10])
    axes[1,0].hist(df_cate_adj['z_scores'], bins=50, density=True, range=[-10,10])
    axes[1,1].hist(df_gcate['z_scores'], bins=50, density=True, range=[-10,10])

    methods = ['GLM-oracle', 'GLM-naive', 'CATE', 'CATE-mad', 'GCATE']
    sns.barplot(data=
                pd.DataFrame({'Methods':methods[1:], 'FDP':np.sum(FP, axis=0) / np.sum(TP + FP, axis=0)})
                , x="Methods", y="FDP", ax=axes[1,2])
    axes[1,2].bar_label(axes[1,2].containers[0], fmt='%.02f')
    for bar, color in zip(axes[1,2].containers[0], sns.color_palette()[:4]):
        bar.set_facecolor(color)

    dict_disc[cell_type] = discoveries

    x_axis = np.arange(-10, 10, 0.001)
    for i in range(2):
        for j in range(3):
            if i==1 and j==2:
                break        
            axes[i,j].plot(x_axis, sp.stats.norm.pdf(x_axis,0,1))
            axes[i,j].set_title(methods[3*i+j])
            axes[i,j].set_ylim([0,0.45])
            if j>0:
                axes[i,j].set_yticks([])
            if i==0:
                axes[i,j].set_xticks([])

    axes[1,2].set_yticks([])
    axes[1,2].set_ylabel('')
    axes[1,2].set_title('FDP')
    axes[1,2].set_ylim([0,1.1])
    axes[1,2].set_xticklabels(axes[1,2].get_xticklabels(), rotation=15, ha='center')
    fig.tight_layout()
    plt.savefig('result/lupus_zscores_{}.pdf'.format(cell_type), bbox_inches='tight', pad_inches=0, dpi=300)
```

```python
stat = []
for cell_type in cell_types:
    arr = dict_disc[cell_type]
    num = arr.shape[0]
    TP = (arr[:,1:] == 1) & (arr[:,:1] == 1)
    FP = (arr[:,1:] == 1) & (arr[:,:1] == 0)
    TN = ((arr[:,1:] == 0) & (arr[:,:1] == 0))
    FN = ((arr[:,1:] == 0) & (arr[:,:1] == 1))
    
    stat.append(
        np.c_[
            np.sum(TP, axis=0)/np.sum(TP+FP, axis=0), 
            np.sum(TN, axis=0)/np.sum(TN+FP, axis=0)
        ]
    )
        
stat = np.stack(stat, axis=0)
df_stat = pd.DataFrame()
for j, metric in enumerate(['Precision', 'Specificity']):
    
    _df = pd.DataFrame(stat[:,:,j], columns=methods[1:])
    _df['Metric'] = metric
    _df['Cell type'] = cell_types
    df_stat = pd.concat([df_stat, _df], axis=0)


df_stat = df_stat.set_index(['Cell type', 'Metric'])

df_stat = df_stat.stack().reset_index()
df_stat.columns = ['Cell type', 'Metric', 'Method', 'Value']
```

```python
from gcate.likelihood import log_h, nll

r_dict = {}
cell_type = 'T4'
file_name = path_data + 'data_lupus_{}.npz'.format(cell_type)
with open(file_name, 'rb') as f:
    tmp = np.load(f, allow_pickle=True)
    X = tmp['X']
    Y = tmp['Y'].astype(float)
    gene_names = tmp['gene_names'].astype(str)
    X = X * 2 - 1
    X = np.c_[np.ones(X.shape[0]), 
                np.log(np.sum(Y, axis=1)),  X]
    X = X / np.sqrt(np.sum(X**2, 0, keepdims=True)) * np.sqrt(X.shape[0])
    id_genes = np.any(Y!=0, axis=0) & (np.sum(Y>0, axis=0)>=10)
    Y = Y[:,id_genes]
    gene_names = gene_names[id_genes]

    
X_sub = np.c_[X[:,:2],X[:,-1:]]
n, p = Y.shape
d = X_sub.shape[1]


with open(path_result+'dill_{}.pkl'.format(cell_type), "rb") as file_object:
    [
        B_glm,tvals,pvals,_df_glm,disp_glm,
        B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
        B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
        B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
    ] = dill.load(file_object)

df_glm_nb = pd.DataFrame({
    'beta_hat':B_glm_nb[:,-1],
    'z_scores':tvals_glm_nb, 
    'p_values':pvals_glm_nb,
    'q_values':multipletests(pvals_glm_nb, alpha=0.05, method='fdr_bh')[1]
})
df_glm_nb_sub = pd.DataFrame({
    'beta_hat':B_glm_nb_sub[:,-1],
    'z_scores':tvals_glm_nb_sub, 
    'p_values':pvals_glm_nb_sub,
    'q_values':multipletests(pvals_glm_nb_sub, alpha=0.05, method='fdr_bh')[1]
})


with open('result/lupus/dill_{}.pkl'.format(cell_type), "rb") as file_object:
    [
        B_glm,tvals,pvals,_df_glm,disp_glm,
        B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
        B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
        B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
    ] = dill.load(file_object)
    
    
kwargs_glm = {
    'family':'negative_binomial',
    'nuisance':1/disp_glm_sub[None,:]
}

df_stat = []
for r in range(1,11):
    with open(path_result+'dill_{}_gcate_r_{}.pkl'.format(cell_type, r), "rb") as file_object:
        [
            A01, A02, A1, A2, P_Gamma, B_de, se
        ] = dill.load(file_object) 
    
    logh = log_h(Y, kwargs_glm['family'], kwargs_glm['nuisance'])
    
    deviance = 2 * ( 
        nll(Y, A1, A2, kwargs_glm['family'], kwargs_glm['nuisance']) / p 
        - np.sum(logh) / (n*p) ) 
    nu = (d + r) * np.maximum(n,p) * np.log(n * p / np.maximum(n,p)) / (n*p)
    jic = deviance + 0.25 * nu
    
    z_scores = (B_de[:,0] / se[:,0])
    z_scores[np.isnan(z_scores)] = np.nanmedian(z_scores)
    pvals = sp.stats.norm.sf(np.abs(z_scores))*2
    df_gcate = pd.DataFrame({
        'z_scores':z_scores, 
        'p_values':pvals,
        'q_values':multipletests(pvals, alpha=0.05, method='fdr_bh')[1]
    })

    df_stat.append([r, np.mean(df_gcate['z_scores']), np.median(df_gcate['z_scores']), np.std(df_gcate['z_scores']),
        sp.stats.median_abs_deviation(z_scores, scale="normal", nan_policy='omit'), 
        sp.stats.skew(z_scores), np.sum(df_gcate['q_values']<0.2), deviance, jic
    ])
    

print(pd.DataFrame(df_stat, columns=['r', 'mean', 'median', 'std', 
                               'mad', 'skewness', 'num_sig', 'deviance', 'JIC'])[
    ['r', 'mean', 'median', 'mad', 'num_sig', 'deviance', 'JIC']].to_latex(
    index=False, float_format="{:.3f}".format,
)  )
```

```python
from gcate.likelihood import log_h, nll

r_dict = {}
cell_type = 'T4'
file_name = path_data + 'data_lupus_{}.npz'.format(cell_type)
with open(file_name, 'rb') as f:
    tmp = np.load(f)
    X = tmp['X']
    Y = tmp['Y'].astype(float)
    X = X * 2 - 1
    X = np.c_[np.ones(X.shape[0]), np.log(np.sum(Y, axis=1)), X]
    X = X / np.sqrt(np.sum(X**2, 0, keepdims=True)) * np.sqrt(X.shape[0])

id_genes = np.any(Y, axis=0) & (np.sum(Y>0, axis=0)>=10)
p_all = Y.shape[1]
Y = Y[:,id_genes]
print(X.shape, Y.shape)

n, p = Y.shape
d = X.shape[1]


with open(path_result+'dill_{}.pkl'.format(cell_type), "rb") as file_object:
    [
        B_glm,tvals,pvals,_df_glm,disp_glm,
        B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
        B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
        B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
    ] = dill.load(file_object)

df_glm_nb = pd.DataFrame({
    'beta_hat':B_glm_nb[:,-1],
    'z_scores':tvals_glm_nb, 
    'p_values':pvals_glm_nb,
    'q_values':multipletests(pvals_glm_nb, alpha=0.05, method='fdr_bh')[1]
})
df_glm_nb_sub = pd.DataFrame({
    'beta_hat':B_glm_nb_sub[:,-1],
    'z_scores':tvals_glm_nb_sub, 
    'p_values':pvals_glm_nb_sub,
    'q_values':multipletests(pvals_glm_nb_sub, alpha=0.05, method='fdr_bh')[1]
})


with open('result/lupus/dill_{}.pkl'.format(cell_type), "rb") as file_object:
    [
        B_glm,tvals,pvals,_df_glm,disp_glm,
        B_glm_nb,tvals_glm_nb,pvals_glm_nb,_df_glm_nb,
        B_glm_sub,tvals_sub,pvals_sub,_df_glm_sub,disp_glm_sub,
        B_glm_nb_sub,tvals_glm_nb_sub,pvals_glm_nb_sub,_df_glm_nb_sub
    ] = dill.load(file_object)
    
    
kwargs_glm = {
    'family':'negative_binomial',
    'nuisance':1/disp_glm[None,:]
}

df_stat = []
for r in range(1,11):
    with open(path_result+'dill_{}_gcate_full_r_{}.pkl'.format(cell_type, r), "rb") as file_object:
        [
            A01, A02, A1, A2, P_Gamma, B_de, se
        ] = dill.load(file_object) 
    
    logh = log_h(Y, kwargs_glm['family'], kwargs_glm['nuisance'])
    
    deviance = 2 * ( 
        nll(Y, A1, A2, kwargs_glm['family'], kwargs_glm['nuisance']) / p 
        - np.sum(logh) / (n*p) ) 
    nu = (d + r) * np.maximum(n,p) * np.log(n * p / np.maximum(n,p)) / (n*p)
    jic = deviance + 0.5 * nu
    
    z_scores = (B_de[:,0] / se[:,0])
    z_scores[np.isnan(z_scores)] = np.nanmedian(z_scores)
    pvals = sp.stats.norm.sf(np.abs(z_scores))*2
    df_gcate = pd.DataFrame({
        'z_scores':z_scores, 
        'p_values':pvals,
        'q_values':multipletests(pvals, alpha=0.05, method='fdr_bh')[1]
    })

    df_stat.append([r, np.mean(df_gcate['z_scores']), np.median(df_gcate['z_scores']), np.std(df_gcate['z_scores']),
        sp.stats.median_abs_deviation(z_scores, scale="normal", nan_policy='omit'), 
        sp.stats.skew(z_scores), np.sum(df_gcate['q_values']<0.2), deviance, jic
    ])
    

print(pd.DataFrame(df_stat, columns=['r', 'mean', 'median', 'std', 
                               'mad', 'skewness', 'num_sig', 'deviance', 'JIC'])[
    ['r', 'mean', 'median', 'mad', 'num_sig', 'deviance', 'JIC']].to_latex(
    index=False, float_format="{:.3f}".format,
)  )
```
